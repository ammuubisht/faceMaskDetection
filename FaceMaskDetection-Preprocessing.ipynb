```python
import cv2, os 
import numpy as np
from keras.utils import np_utils
```

```python
data_path = 'C:/face_mask_dataset'
categories = os.listdir(data_path)
labels = [label for label in range(len(categories))]

label_dict = dict(zip(categories, labels))

data = []
target = []

for category in categories:
    in_data_path = os.path.join(data_path, category)
    imgs = os.listdir(in_data_path)
    
    for img in imgs:
        img_path = os.path.join(in_data_path, img)
        img = cv2.imread(img_path)
        
        try: 
            gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            resized_img = cv2.resize(gray_img, (100,100))
            data.append(resized_img)
            target.append(label_dict[category])
            
        except Exception as e:
            print("Exception: ", e)
            
```

```python
data = np.array(data)/255.0
data = np.reshape(data, (data.shape[0], 100, 100, 1))
target = np.array(target)
data[:5]
```

```python
new_target = np_utils.to_categorical(target)
np.save('data', data)
np.save('target', new_target)
```
